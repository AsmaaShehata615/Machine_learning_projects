import dependencies 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data collection and analysis

```python
data = pd.read_csv('/content/diabetes.csv')
```

```python
data.head(5)
```

0--> non diabetic

1--> diabetic

```python
data.groupby('Outcome').mean()
```

```python
x= data.drop(columns= 'Outcome', axis= 1)
y= data['Outcome']
```

```python
  #data standarization 

  scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data= scaler.transform(x)
```

```python
print(standardized_data)
```

```python
Xtrain, Xtest, ytrain , ytest = train_test_split(standardized_data, y , test_size=0.2, stratify=y, random_state= 1)
```

```python
 classifier = svm.SVC(kernel= 'linear')
```

```python
#training the support vector machine
classifier.fit(Xtrain, ytrain)
```

model evalution 

accuracy score


```python
#accuracy score 
training_prediction= classifier.predict(Xtrain)
```

```python
training_accuracy=accuracy_score(training_prediction,ytrain)
```

```python
print('accuracy of training set', training_accuracy)
```

```python
#accuracy score 
testing_prediction= classifier.predict(Xtest)
```

```python
testing_accuracy=accuracy_score(testing_prediction,ytest)
```

```python
print('accuracy of testing set', testing_accuracy)
```
